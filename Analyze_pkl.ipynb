```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

```

## 1. Analyse seed.pkl

```python
with open('240523_01_seeding.pkl', 'rb') as f:
    data = pickle.load(f)

print(data)
```

```python
X_pos = data["X_pos"]
X_pos_measured = data["X_pos_measured"]
X_pos_unmeasured = data["X_pos_unmeasured"]
plt.figure(figsize=(10, 5))
plt.scatter(X_pos, np.ones_like(X_pos), color='blue', label='All Positions')
plt.scatter(X_pos_measured, np.ones_like(X_pos_measured)*2, color='green', label='Measured Positions')
plt.scatter(X_pos_unmeasured, np.ones_like(X_pos_unmeasured)*3, color='red', label='Unmeasured Positions')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.title("Positions_plot")
plt.show()
```

```python
indices_all = data["indices_all"]
indices_measured = data["indices_measured"]
indices_unmeasured = data["indices_unmeasured"]
plt.figure(figsize=(10, 5))
plt.scatter(indices_all, np.ones_like(indices_all), color='blue', label='All Indices')
plt.scatter(indices_measured, np.ones_like(indices_measured)*2, color='green', label='Measured Indices')
plt.scatter(indices_unmeasured, np.ones_like(indices_unmeasured)*3, color='red', label='Unmeasured Indices')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.title("Indices_plot")
plt.show()
```

```python
num_images = len(data["X_topo_measured"])
print(num_images)
grid_size = int(np.ceil(np.sqrt(num_images)))

plt.figure(figsize=(10, 10))
plt.title("Seed")
for i in range(num_images):
    plt.subplot(grid_size, grid_size, i+1)
    plt.imshow(data["X_topo_measured"][i])
    plt.axis('off')

plt.show()
```

```python
num_images = len(data["X_topo"])
print(num_images)
grid_size = int(np.ceil(np.sqrt(num_images)))

plt.figure(figsize=(10, 10))

for i in range(num_images):
    plt.subplot(grid_size, grid_size, i+1)
    plt.imshow(data["X_topo"][i])
    plt.axis('off')

plt.show()
```
